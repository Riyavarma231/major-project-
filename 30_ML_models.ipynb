```python
import lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
import pandas as pd 

data = pd.read_csv("/content/Breast_Data.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Target', axis = 1)
Y = data['Target'].copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
#Encoding y class label
Y = Y.map({"Breast": 1, "Negative": 0})  
```

```python
X = sc.fit_transform(X)
```

```python
# data split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state =42, stratify=Y)
```

```python
Y_test
```

```python
from sklearn.metrics import matthews_corrcoef
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=matthews_corrcoef)
```

```python
#model test
models,predictions = clf.fit(X_train, X_test, Y_train, Y_test)
```

```python
print(models)
```

```python
# Plot of Accuracy
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=models.index, x="Accuracy", data=models)
ax.set(xlim=(0, 1))
```

```python
# Plot of MCC
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=models.index, x="matthews_corrcoef", data=models)
ax.set(xlim=(0, 1))
```
