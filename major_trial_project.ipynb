Amino Acid Composition

```python
# Amino acid composition (AAC)

from Pfeature.pfeature import aac_wp

def aac(input):
  a = input.rstrip('txt')
  output = a + 'aac.csv'
  df_out = aac_wp(input, output)
  df_in = pd.read_csv(output)
  return df_in

aac('/content/Breast_train_main_cdhit.txt')
```

```python
! ls -l
```

```python
#Calculate feature for each cancer data + combining the data + merge with cancer type

Bre = '/content/Breast_train_main_cdhit.txt'
Cer = '/content/Cervix_train_main_cdhit.txt'
Col = '/content/Colon_train_main_cdhit.txt'
Lun = '/content/Lungs_train_main_cdhit.txt'
Pro = '/content/Prostate_train_main_cdhit.txt'
Ski = '/content/skin_train_main_cdhit.txt'
neg = '/content/neg_train_main_cdhit.txt'

def feature_calc(Bre, Cer, Col, Lun, Pro, Ski, neg, feature_name):
  # Calculate cancer feature
  Bre_feature = feature_name(Bre)
  Cer_feature = feature_name(Cer)
  Col_feature = feature_name(Col)
  Lun_feature = feature_name(Lun)
  Pro_feature = feature_name(Pro)
  Ski_feature = feature_name(Ski)
  neg_feature = feature_name(neg)
  
  Bre_Cer_Col_Lun_Pro_Ski_neg_feature = pd.concat([Bre_feature, Cer_feature, Col_feature, Lun_feature, Pro_feature, Ski_feature, neg_feature], axis=0)
 
  # Combine feature
  df = pd.concat([Bre_feature, Cer_feature, Col_feature, Lun_feature, Pro_feature, Ski_feature, neg_feature], join="inner")
  return df

feature = feature_calc(Bre, Cer, Col, Lun, Pro, Ski, neg, aac) # AAC
feature
```
